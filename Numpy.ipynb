```python
#NumPy Copy vs View
#if original changes copy doesnt
import numpy as np

arr=np.array([1,2,3])

x=arr.copy()

arr[0]=100

print(arr)
print(x)
```

```python
#View
#it is linked with original array
arr=np.array([1,2,3])

x=arr.view()

arr[0]=100

print(arr)
print(x)
```

```python
#nditer() (Professional Method)
import numpy as np

arr = np.array([
    [1,2],
    [3,4]
])

for i in np.nditer(arr):#NumPy built-in iterator is nditer() , which  accesss each element of multi-dimensional arrays
    print(i)
```

```python
import numpy as np

arr1 = np.array([
    [1,2],
    [3,4]
])

arr2 = np.array([
    [5,6],
    [7,8]
])

print(np.concatenate((arr1, arr2), axis=1))#axis 1 means join columns
```

```python
#Horizontal Stack also used for join
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.hstack((a,b)))
```

```python
#Vertical Stack for join
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.vstack((a,b)))
```

```python
#Depth Stack for join
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.dstack((a,b)))
```

```python
#NumPy Array Join
import numpy as np

a = np.array([10,20,30])
b = np.array([40,50,60])

# 1. Har element print karo
for i in a:
    print(i)

# 2. Dono arrays ko join karo
c = np.concatenate((a,b))
print(c)

# 3. Vertical stack
print(np.vstack((a,b)))

# 4. Horizontal stack
print(np.hstack((a,b)))
```

```python
#complete practice code
import numpy as np

# Create Array
arr = np.array([10,20,30,40,50,60])

print("Original:", arr)

# Indexing
print("First:", arr[0])

# Slicing
print("Slice:", arr[1:4])

# Data Type
print("Datatype:", arr.dtype)

# Reshape
new = arr.reshape(2,3)
print("Reshape:\n", new)

# Shape
print("Shape:", new.shape)

# Iteration
print("Loop:")
for i in arr:
    print(i)

# Join
a=np.array([1,2])
b=np.array([3,4])
print("Join:",np.concatenate((a,b)))

# Split
print("Split:",np.array_split(arr,3))

# Search
print("Search:",np.where(arr==40))

# Sort
x=np.array([5,3,1,4,2])
print("Sort:",np.sort(x))

# Filter
print("Greater than 30:",arr[arr>30])
```
