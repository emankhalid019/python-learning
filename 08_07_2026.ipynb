```python
# Python Try Except

print(" Basic Try Except ")

try:
    # Risky code (Variable does not exist)
    print(x)

except:
    # Handle any error
    print("Error: Variable x is not defined")

print("\n Multiple Exceptions ")

try:
    # Divide by zero (will raise ZeroDivisionError)
    result = 10 / 0
    print(result)

except ZeroDivisionError:
    # Handle divide by zero error
    print("Cannot divide by zero")

except NameError:
    # Handle variable error
    print("Variable not found")

except:
    # Handle all other errors
    print("Unknown error")

print("\n===== Example 3 : Else Block =====")

try:
    # No error here
    number = 20
    print("Number =", number)

except:
    print("Something went wrong")

else:
    # Runs only if no error occurs
    print("No Error Found")

print("\n Finally Block ")

try:
    # Risky code
    print(50 / 5)

except:
    print("Error")

finally:
    # Always runs
    print("Program Finished")

print("\n File Handling")

try:
    # Open file in read mode
    file = open("sample.txt", "r")

    try:
        # Read file content
        print(file.read())

    except:
        # Handle reading error
        print("Error while reading file")

    finally:
        # Close file
        file.close()
        print("File Closed")

except:
    # Handle file not found error
    print("File does not exist")

print("\nProgram Ended Successfully")
```

```python
# Python String Formatting

# Variables
name = "Eman"
age = 22
price = 99.5678
salary = 500000
marks = 0.95
tax = 20

# Basic F-String
print(" Basic F-String ")

# Insert variable inside string
print(f"My name is {name}")
print(f"My age is {age}")

print("\n Decimal Formatting ")

# Show only 2 decimal places
print(f"Price = {price:.2f}")

print("\n Operations")

# Perform calculations inside placeholder
print(f"20 + 30 = {20+30}")
print(f"20 × 5 = {20*5}")
print(f"Price + Tax = {price+tax}")

# If Else in Placeholder

print("\n If Else ")

# Display Expensive or Cheap
print(f"Product is {'Expensive' if price>50 else 'Cheap'}")

print("\n Functions ")

# Built-in function
print(f"Upper = {name.upper()}")

# User-defined function
def square(x):
    return x*x

print(f"Square of 6 = {square(6)}")

print("\n Comma Separator ")

print(f"Salary = {salary:,}")

print("\n  Percentage ")

print(f"Marks = {marks:.0%}")

print("\n Binary Octal Hex")

number = 255

print(f"Binary = {number:b}")
print(f"Octal = {number:o}")
print(f"Hex = {number:x}")

print("\n Alignment")

print(f"|{'Python':<15}|")   # Left Align
print(f"|{'Python':>15}|")   # Right Align
print(f"|{'Python':^15}|")   # Center Align

# Old format() Method
print("\n  format() Method ")

print("Hello {}".format(name))
print("{} is {} years old".format(name, age))

# Index Numbers
print("{1} scored {0}%".format(95, name))

# Named Index
print("Welcome {student}".format(student=name))
```

```python
# Python None Complete Example

print(" Assign None")

# Variable has no value yet
student = None

print("Student =", student)

print("\n Data Type")

# Check data type
print("Type =", type(student))

print("\n Compare with None ")

# Check if value is None
if student is None:
    print("Student data is not available")
else:
    print("Student data found")

print("\n Assign Value ")

# Assign actual value
student = "Eman"

print("Student =", student)

print("\n is not None ")

# Check if value exists
if student is not None:
    print("Student data is available")
else:
    print("No student data")

print("\n Difference ")

print("None == 0      :", None == 0)
print("None == False  :", None == False)
print("None == ''     :", None == "")
print("None is None   :", None is None)

print("\n Real-Life Example ")

# Initially no marks available
marks = None

if marks is None:
    print("Marks have not been uploaded yet.")

# Teacher uploads marks
marks = 95

if marks is not None:
    print(f"Student Marks = {marks}")

print("\nProgram Finished Successfully")
```

```python
# Python User Input Complete Example
import math

print(" Simple Input ")

# Ask user for name
name = input("Enter your name: ")

# Display user name
print(f"Hello {name}")

print("\n Multiple Inputs")

# Take multiple inputs
city = input("Enter your city: ")
color = input("Favorite color: ")

# Display inputs
print(f"{name} lives in {city}.")
print(f"Favorite Color = {color}")

print("\n Input is String ")

# Input is always string
age = input("Enter your age: ")

print("Value =", age)
print("Type =", type(age))

print("\n Convert to Integer")

# Convert string into integer
age = int(input("Enter your age again: "))

print("Age after 5 years =", age + 5)

print("\n Convert to Float")

# Convert string into float
price = float(input("Enter product price: "))

print(f"Price = {price:.2f}")

print("\n Square Root ")

# Take number
number = float(input("Enter a number: "))

# Calculate square root
print(f"Square Root = {math.sqrt(number):.2f}")

print("\n Input Validation ")

# Keep asking until valid number entered
while True:

    try:
        marks = float(input("Enter your marks: "))

        # Exit loop if input is valid
        break

    except:
        print("Invalid Input! Please enter numbers only.")

print(f"Marks = {marks}")

print("\n Simple Calculator")

# Take two numbers
num1 = float(input("Enter First Number: "))
num2 = float(input("Enter Second Number: "))

# Display calculations
print(f"Addition = {num1 + num2}")
print(f"Subtraction = {num1 - num2}")
print(f"Multiplication = {num1 * num2}")

# Avoid divide by zero
if num2 != 0:
    print(f"Division = {num1 / num2:.2f}")
else:
    print("Cannot divide by zero.")

print("\nProgram Finished Successfully")
```

```python
# Python Classes and Objects
print(" Create Class ")

# Create a simple class
class MyClass:

    # Class Property
    x = 5

print("\n Create Object ")

# Create object from class
p1 = MyClass()

# Access property
print("Value of x =", p1.x)

print("\n Multiple Objects ")

# Create multiple objects
p2 = MyClass()
p3 = MyClass()

print("p1 =", p1.x)
print("p2 =", p2.x)
print("p3 =", p3.x)

print("\n Independent Objects")

# Another class
class Student:

    name = "Unknown"
    cgpa = 3.0

# Create objects
s1 = Student()
s2 = Student()

# Modify only s1
s1.name = "Baidar"
s1.cgpa = 3.50

# Modify only s2
s2.name = "Ali"
s2.cgpa = 3.20

print("Student 1:", s1.name, s1.cgpa)
print("Student 2:", s2.name, s2.cgpa)

print("\n Delete Object ")

# Create object
temp = MyClass()

print("Before Delete =", temp.x)

# Delete object
del temp

print("Object deleted successfully.")

print("\n Empty Class")

# Empty class
class Person:
    pass

# Create object
person1 = Person()

print("Empty class object created successfully")
```

```python
# Python __init__()

print(" Simple Constructor")

# Create a class
class Person:

    # Constructor runs automatically
    def __init__(self, name, age):

        # Store values in object
        self.name = name
        self.age = age

# Create object
p1 = Person("Baidar", 22)

# Display values
print("Name :", p1.name)
print("Age  :", p1.age)

print("\n= Multiple Objects")

p2 = Person("Ali", 20)

p3 = Person("Ahmed", 24)

print(p2.name, p2.age)
print(p3.name, p3.age)

print("\n Without __init__()")

class Student:
    pass

# Create object
s1 = Student()

# Manually assign values
s1.name = "Ayesha"
s1.cgpa = 3.50

print("Name :", s1.name)
print("CGPA :", s1.cgpa)

print("\n Default Value ")

class Employee:

    # Default salary = 30000
    def __init__(self, name, salary=30000):

        self.name = name
        self.salary = salary

e1 = Employee("Sara")

e2 = Employee("Ayesha",50000)

print(e1.name, e1.salary)

print(e2.name, e2.salary)

print("\n Multiple Parameters")

class Teacher:

    def __init__(self,name,age,city,country):

        self.name=name

        self.age=age

        self.city=city

        self.country=country

t1 = Teacher(
    "Usman",
    35,
    "Lahore",
    "Pakistan"
)

print("Name    :",t1.name)

print("Age     :",t1.age)

print("City    :",t1.city)

print("Country :",t1.country)

print("\n Another Student ")

class UniversityStudent:

    def __init__(self,name,semester,cgpa):

        self.name=name

        self.semester=semester

        self.cgpa=cgpa

student1 = UniversityStudent(
    "Eman",
    6,
    3.06
)

print("Student :",student1.name)

print("Semester:",student1.semester)

print("CGPA    :",student1.cgpa)
```

```python
# Python self Parameter(Current object reference which is use to access Object properties and Object methods)

# Create a class
class Student:

    # Constructor (__init__) runs automatically when object is created
    def __init__(self, name, age, department):

        # Store values inside current object
        self.name = name
        self.age = age
        self.department = department

    # Method to display student information
    def display(self):
        print("\n Student Information ")
        print(f"Name       : {self.name}")
        print(f"Age        : {self.age}")
        print(f"Department : {self.department}")

    # Method that returns greeting
    def greet(self):
        return f"Hello {self.name}"

    # Calling another method using self
    def welcome(self):
        message = self.greet()      # Call greet() using self
        print(message)
        print("Welcome to Python Learning!\n")

# Create multiple objects

student1 = Student("Ali", 21, "Computer Science")
student2 = Student("Ahmed", 22, "Artificial Intelligence")

# Call methods for first student

student1.display()
student1.welcome()

# Call methods for second student

student2.display()
student2.welcome()

# Access object properties directly

print("Direct Access:")
print(student1.name)
print(student2.department)
```
