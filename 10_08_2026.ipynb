```python
# Python File Handling

# 1. Create a new file (x mode)
# Creates file only if it does not already exist
try:
    file = open("student.txt", "x")
    file.write("Eman\n")
    file.close()
    print("File Created Successfully!")
except FileExistsError:
    print("File already exists!")

# 2. Write Mode (w)
# Overwrites existing data
file = open("student.txt", "w")
file.write("Eman\n")
file.write("Ali\n")
file.write("Ahmed\n")
file.close()
print("\nData Written Successfully!")

# 3. Append Mode (a)
# Adds new data at the end
file = open("student.txt", "a")
file.write("Sara\n")
file.close()
print("New Data Appended!")

# 4. Read Mode (r)
# Reads complete file
file = open("student.txt", "r")

print("\nReading File")
print(file.read())

file.close()

# 5. Default Mode
# Same as open("student.txt","rt")
file = open("student.txt")

print("Reading Again")
print(file.read())
file.close()
```

```python
# Python File Open & Read Example

# NOTE:
# Make sure "student.txt" exists in the same folder
# Example content:
# Eman
# Ali
# Ahmed
# Sara

# 1. Open file and read complete file
file = open("student.txt", "r")   # Open file in read mode

print("Complete File:\n")
print(file.read())                # Read complete file

file.close()                      # Close file

# 2. Read only first 10 characters
with open("student.txt", "r") as file:

    print("\nFirst 10 Characters:")
    print(file.read(10))

# File automatically closes here

# 3. Read one line
with open("student.txt", "r") as file:

    print("\nFirst Line:")
    print(file.readline())

# 4. Read first two lines
with open("student.txt", "r") as file:

    print("\nFirst Two Lines:")
    print(file.readline())
    print(file.readline())

# 5. Read file line by line using loop
with open("student.txt", "r") as file:

    print("\nReading Line by Line:")

    for line in file:
        print(line.strip())   # strip() removes extra newline
```

```python
# Python File Write Example

# 1. Create a new file using x mode
# (Runs only if file does not already exist)

try:
    file = open("student.txt", "x")
    file.close()
    print("File Created Successfully!")
except FileExistsError:
    print("File already exists!")

# 2. Write Mode (w)
# Deletes old content and writes new data
with open("student.txt", "w") as file:
    file.write("Eman\n")
    file.write("Ali\n")
    file.write("Ahmed\n")

print("\nData Written Successfully!")

# Read file
with open("student.txt", "r") as file:

    print("\nCurrent File Content:")
    print(file.read())

# 3. Append Mode (a)
# Adds new data at the end
with open("student.txt", "a") as file:

    file.write("Sara\n")
    file.write("Hamza\n")
print("\nNew Data Appended!")

# Read file again
with open("student.txt", "r") as file:

    print("\nUpdated File Content:")
    print(file.read())
```

```python
# Python Delete File Example

# Import os module
# Used for file and folder operations
import os

# 1. Check if file exists

if os.path.exists("student.txt"):

    # Delete the file
    os.remove("student.txt")
    print("File Deleted Successfully!")

else:
    print("File does not exist.")

# 2. Delete an Empty Folder

folder_name = "myfolder"

if os.path.exists(folder_name):
    try:
        # Deletes only an EMPTY folder
        os.rmdir(folder_name)

        print("Folder Deleted Successfully!")

    except OSError:
        print("Folder is not empty. Cannot delete.")

else:
    print("Folder does not exist.")
```

```python
# Python Complete File Handling Example

# Import OS module Used for checking, deleting files
import os
file_name = "students.txt"

# 1. Create File (if not exists)
# "w" mode creates file automatically if it doesn't exist
with open(file_name, "w") as file:

    file.write("Name\tAge\tDepartment\n")
    file.write("Eman\t20\tComputer Science\n")
    file.write("Ali\t21\tSoftware Engineering\n")
print("File Created and Data Written Successfully!")

# 2. Read Complete File
print("\n Reading Complete File\n")
with open(file_name, "r") as file:

    print(file.read())

# 3. Read Line by Line
print("\n Reading File Line by Line\n")
with open(file_name, "r") as file:

    for line in file:
        print(line.strip())

# 4. Append New Record
with open(file_name, "a") as file:
    file.write("Ahmed\t22\tArtificial Intelligence\n")

print("\nNew Student Added Successfully!")

# 5. Read Updated File
print("\n Updated File Data\n")
with open(file_name, "r") as file:
    print(file.read())

# 6. Delete File
choice = input("\nDo you want to delete the file? (yes/no): ")
if choice.lower() == "yes":

    if os.path.exists(file_name):
        os.remove(file_name)
        print("File Deleted Successfully!")

    else:
        print("File does not exist.")

else:
    print("File is Safe.")
```
