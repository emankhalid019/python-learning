```python
#Python Math
import math

print("Built-in Math Functions")

numbers = [5, 10, 25, -7]

print("Numbers:", numbers)

print("Minimum:", min(numbers))

print("Maximum:", max(numbers))

print("Absolute Value:", abs(-25))

print("Power (4^3):", pow(4,3))

print()

print("Math Module Functions")

print("Square Root of 64:", math.sqrt(64))

print("Square Root of 144:", math.sqrt(144))

print()

print("Ceil of 7.2:", math.ceil(7.2))

print("Ceil of 5.01:", math.ceil(5.01))

print()

print("Floor of 7.9:", math.floor(7.9))

print("Floor of 9.99:", math.floor(9.99))

print()

print("Value of PI:", math.pi)

print()

radius = 7

area = math.pi * radius * radius

print("Radius:", radius)

print("Area of Circle:", area)

print()

print("")

print("Power (2^5):", pow(2,5))

print("Absolute Value:", abs(-100))

print("Maximum:", max(50,90,10,35))

print("Minimum:", min(50,90,10,35))
```

```python
#Python JSON(text format which is used to store & exchange data)

import json

print(" JSON to Python")

json_data = '{"name":"Ali","age":22,"city":"Lahore"}'

python_data = json.loads(json_data)

print(python_data)

print("Name:", python_data["name"])

print("Age:", python_data["age"])

print()

print("Python to JSON")

student = {

    "name":"Ahmed",

    "age":21,

    "married":False,

    "subjects":["Python","Java","C++"],

    "marks":{

        "Python":90,

        "Java":85

    },

    "hostel":None

}

json_string = json.dumps(student)

print(json_string)

print()

print("Pretty JSON")

print(json.dumps(student,indent=4))

print()

print("Sorted Keys")

print(json.dumps(student,indent=4,sort_keys=True))

print()

print("Different Python Objects")

print(json.dumps(["Apple","Banana"]))

print(json.dumps(("Python","Java")))

print(json.dumps(True))

print(json.dumps(False))

print(json.dumps(None))

print(json.dumps(100))

print(json.dumps(25.5))

print()

print(" Custom Separators")

print(json.dumps(student,indent=4,separators=(". "," = ")))
```

```python
# Import JSON module
# Used to convert data between Python and JSON
import json

# JSON String in Python Dictionary (loads)

# JSON data stored as a string
json_data = '{"name":"Ali","age":22,"city":"Lahore"}'

# Convert JSON string into Python dictionary
python_data = json.loads(json_data)

print(" JSON to Python")
print(python_data)
print("Name:", python_data["name"])
print("Age:", python_data["age"])
print("City:", python_data["city"])

# Python Dictionary in JSON String (dumps)
# Python dictionary
student = {
    "name": "Ahmed",
    "age": 21,
    "city": "Sahiwal",
    "married": False,
    "subjects": ["Python", "Java", "C++"],
    "marks": {
        "Python": 90,
        "Java": 85
    },
    "hostel": None
}

# Convert Python dictionary into JSON string
json_string = json.dumps(student)

print("\n Python to JSON")
print(json_string)
# Pretty JSON (indent)
# Makes JSON easy to read
print("\nPretty JSON")
print(json.dumps(student, indent=4))

# Sort Keys
# Display keys in alphabetical order
print("\n Sorted Keys ")
print(json.dumps(student, indent=4, sort_keys=True))

# Convert Different Python Objects into JSON
print("\n  Different Data Types")

# Dictionary -> JSON Object
print(json.dumps({"course": "Python"}))

# List in JSON Array
print(json.dumps(["Apple", "Banana", "Orange"]))

# Tuple in JSON Array
print(json.dumps(("Red", "Green", "Blue")))

# String in JSON String
print(json.dumps("Hello World"))

# Integer in JSON Number
print(json.dumps(100))

# Float in JSON Number
print(json.dumps(45.75))

# Boolean in JSON Boolean
print(json.dumps(True))
print(json.dumps(False))

# None in JSON null
print(json.dumps(None))
# Custom Separators
# Change default comma and colon
print("\n Custom Separators")
print(json.dumps(student, indent=4, separators=(" | ", " = ")))
```

```python
#RegEx(Regular Expression)use to search character pattern in a string
import re

# Sample text
text = "The rain in Spain 2026"
print("Original Text:")
print(text)

# Find all matches

# Find every occurrence of "ai"
print("\nFind All:")
print(re.findall("ai", text))

# Search first match

# Search first occurrence
match = re.search("Spain", text)

print("\nSearch:")
print(match)

# Print matched word
print("Matched Word:", match.group())

# Print starting position
print("Start Position:", match.start())

# Print span (start,end)
print("Span:", match.span())

# Split String

# Split text at every space
print("\nSplit:")
print(re.split(r"\s", text))

# Replace Text
# Replace spaces with "-"
print("\nReplace:")
print(re.sub(r"\s", "-", text))

# ^ Starts With
print("\nStarts With The?")
print(re.search("^The", text))

# $ Ends With 2026
print("\nEnds With 2026?")
print(re.search("2026$", text))

# Digits
print("\nDigits:")
print(re.findall(r"\d", text))

# Words
print("\nWord Characters:")
print(re.findall(r"\w", text))

# Capital Letters
print("\nCapital Letters:")
print(re.findall(r"[A-Z]", text))

# Lower Case Letters
print("\nSmall Letters:")
print(re.findall(r"[a-z]", text))

# Ignore Case
print("\nIgnore Case:")
print(re.search("spain", text, re.I))
```
