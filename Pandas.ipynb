```python
# PANDAS - Intro, Getting Started, Series, DataFrames,Read CSV,Read JSON,Analyzing Data


import pandas as pd
import numpy as np


# Pandas is a Python library used for working with data sets.It has functions for analyzing, cleaning, exploring, and manipulating data.
# The name "Pandas" refers to both "Panel Data" and "Python Data Analysis", and was created by Wes McKinney in 2008.

# What can Pandas do?
# - Discover correlations between columns
# - Calculate statistics (average, min, max, etc.)
# - Clean messy data (remove rows with missing/wrong data)
# - Analyze large data sets and make conclusions

# Check the installed pandas version:
print("Pandas version:", pd.__version__)


## PANDAS GETTING STARTED

# Pandas is usually imported under the alias "pd":
#   import pandas as pd
#
# Once imported, all pandas features are accessed via pd.<something>
# e.g. pd.DataFrame(), pd.Series(), pd.read_csv(), etc.

mydataset = {
    'cars': ["BMW", "Toyota", "Honda"],
    'passings': [3, 7, 2]
}

myvar = pd.DataFrame(mydataset)
print(myvar)


## PANDAS SERIES
# A Series is a one-dimensional array holding data of any type (like a single column in a table).

a = [1, 7, 2]
myseries = pd.Series(a)
print(myseries)
# The left-hand numbers (0, 1, 2...) are automatically created labels, called the INDEX.

# Access a value using its index label
print("Value at index 0:", myseries[0])

# Create labels (custom index) instead of default 0,1,2...
myseries2 = pd.Series(a, index=["x", "y", "z"])
print(myseries2)
print("Value at label 'y':", myseries2["y"])

# Create a Series from a dictionary (keys become the index labels)
calories = {"day1": 420, "day2": 380, "day3": 390}
myseries3 = pd.Series(calories)
print(myseries3)

# Use only some of the items from a dictionary by specifying keys
myseries4 = pd.Series(calories, index=["day1", "day2"])
print(myseries4)


## PANDAS DATAFRAMES
# A DataFrame is a 2-dimensional data structure, like a table with rows and columns. A Pandas Series is like a single column; a DataFrame is the whole table.

data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}
df = pd.DataFrame(data)
print(df)

# Locate a row using loc[] - returns a Series representing that row
print(df.loc[0])

# Locate multiple rows - returns a DataFrame
print(df.loc[[0, 1]])

# Named indexes instead of default 0,1,2...
df2 = pd.DataFrame(data, index=["day1", "day2", "day3"])
print(df2)

# Locate a named row
print(df2.loc["day2"])



## PANDAS READ CSV
# CSV (Comma Separated Values) files store tabular data as plaintext, one row per line, values separated by commas.
# pd.read_csv() loads a CSV file into a DataFrame.

# Create a sample CSV file first (so this cell is self-contained)
csv_data = """StudentID,Name,Age,Department
1,Ali,20,CS
2,Sara,21,IT
3,Ahmed,19,CS
4,Ayesha,22,IT
5,Zarmeen,23,CS
"""
with open("students.csv", "w") as f:
    f.write(csv_data)

df_csv = pd.read_csv('students.csv')
print(df_csv)

# to_string() prints the ENTIRE DataFrame, even if it's very large (by default, pandas only shows the first and last few rows)
print(df_csv.to_string())

# Check/change the maximum number of rows pandas shows by default
print("Default max rows shown:", pd.options.display.max_rows)



## PANDAS READ JSON
# JSON (JavaScript Object Notation) is a common data format,structurally very similar to a Python dictionary.
# pd.read_json() loads JSON data into a DataFrame.

json_data = """
{
  "StudentID": {"0": 1, "1": 2, "2": 3},
  "Name": {"0": "Ali", "1": "Sara", "2": "Ahmed"},
  "Age": {"0": 20, "1": 21, "2": 19}
}
"""
with open("students.json", "w") as f:
    f.write(json_data)

df_json = pd.read_json('students.json')
print(df_json)

# A Python dictionary can also be loaded directly as a DataFrame (this behaves the same as reading JSON, since JSON = dict structure)
data_dict = {
    "Name": ["Ali", "Sara", "Ahmed"],
    "Age": [20, 21, 19]
}
df_from_dict = pd.DataFrame(data_dict)
print(df_from_dict)



## PANDAS ANALYZING DATA
# Pandas provides quick, built-in ways to explore and understand ,a DataFrame without manually inspecting every row.

# head(n) -- returns the first n rows (default 5)
print(df_csv.head())
print(df_csv.head(2))

# tail(n) -- returns the last n rows (default 5)
print(df_csv.tail(2))

# info() -- prints a technical summary: column names, data types,non-null counts, and memory usage
df_csv.info()

# describe() -- shows summary statistics for numeric columns(count, mean, std, min, 25%, 50%, 75%, max)
print(df_csv.describe())

# shape -- returns (number_of_rows, number_of_columns)
print("Shape:", df_csv.shape)

# columns -- lists all column names
print("Columns:", list(df_csv.columns))

# dtypes -- shows the data type of each column
print(df_csv.dtypes)




# Concept              Purpose                                  Example

# pd.Series()            1D labeled array (like one column)        pd.Series([1,7,2])
# pd.DataFrame()          2D labeled table (rows + columns)         pd.DataFrame({"a":[1,2]})
# df.loc[]                Access row(s) by label/index               df.loc[0] / df.loc["day1"]
# pd.read_csv()            Load a CSV file into a DataFrame           pd.read_csv('file.csv')
# pd.read_json()           Load JSON data into a DataFrame            pd.read_json('file.json')
# df.to_string()           Print the full DataFrame                  df.to_string()
# df.head(n)               First n rows                              df.head(3)
# df.tail(n)               Last n rows                               df.tail(3)
# df.info()                Summary of columns/types/nulls             df.info()
# df.describe()            Summary statistics for numeric columns    df.describe()
# df.shape                 (rows, columns) count                     df.shape
```
