```python
#Python range
# Range with one argument
print("One Argument:")
print(list(range(10)))

print()

# Range with two arguments
print("Two Arguments:")
print(list(range(3, 10)))

print()

# Range with three arguments
print("Three Arguments:")
print(list(range(3, 10, 2)))

print()

# Range in for loop
print("For Loop:")
for i in range(5):
    print(i)

print()

# Indexing
r = range(10)

print("Index Value:", r[2])

print()

# Slicing
print("Slice:", list(r[:5]))

print()

# Membership
r = range(0, 10, 2)

print("6 in Range:", 6 in r)

print("7 in Range:", 7 in r)

print()

# Length
print("Length:", len(r))

print()

# Negative Step
print("Reverse Range:")
print(list(range(10, 0, -1)))
```

```python
# Python Arrays (Lists) Practice

# Create array
cars = ["Ford", "Volvo", "BMW"]

print("Original Array:")
print(cars)

print()

# Access element
print("First Car:", cars[0])

print()

# Modify element
cars[0] = "Toyota"

print("After Modification:")
print(cars)

print()

# Length
print("Length:", len(cars))

print()

# Loop through array
print("Looping:")
for car in cars:
    print(car)

print()

# Append element (add value at end)
cars.append("Honda")

print("After Append:")
print(cars)

print()

# Insert element
cars.insert(1, "Audi")

print("After Insert:")
print(cars)

print()

# Extend array
more = ["Kia", "Suzuki"]

cars.extend(more)

print("After Extend:")
print(cars)

print()

# Count
numbers = [1, 2, 2, 3, 2]

print("Count of 2:", numbers.count(2))

print()

# Index
print("Index of Honda:", cars.index("Honda"))

print()

# Copy
copy_cars = cars.copy()

print("Copied Array:")
print(copy_cars)

print()

# Reverse
cars.reverse()

print("After Reverse:")
print(cars)

print()

# Sort
cars.sort()

print("After Sort:")
print(cars)

print()

# Pop
cars.pop(1)

print("After Pop:")
print(cars)

print()

# Remove
cars.remove("Honda")

print("After Remove:")
print(cars)

print()

# Clear copy
copy_cars.clear()

print("After Clear:")
print(copy_cars)
```

```python
# Python Iterators (One item at a time dene wala obj)

# Create an iterable (tuple)
fruits = ("Apple", "Banana", "Mango")

# Convert iterable into iterator
fruit_iterator = iter(fruits)

print("Tuple Iterator:")

print(next(fruit_iterator))
print(next(fruit_iterator))
print(next(fruit_iterator))

print()

# String Iterator
text = "Python"

text_iterator = iter(text)

print("String Iterator:")

print(next(text_iterator))
print(next(text_iterator))
print(next(text_iterator))
print(next(text_iterator))
print(next(text_iterator))
print(next(text_iterator))

print()

# For loop automatically uses iterator
print("Tuple using for loop:")

for fruit in fruits:

    print(fruit)

print()

# Custom Iterator
class MyNumbers:

    def __iter__(self):

        self.number = 1

        return self

    def __next__(self):

        if self.number <= 5:

            x = self.number

            self.number += 1

            return x

        else:

            raise StopIteration

obj = MyNumbers()

iterator = iter(obj)

print("Custom Iterator:")

for value in iterator:

    print(value)
```

```python
#Python module
import types

# Create a module object
mymodule = types.ModuleType("mymodule")

# Function
def greeting(name):
    print("Hello,", name)

# Variables
mymodule.greeting = greeting
mymodule.college = "ABC University"
mymodule.subjects = ["Python", "Java", "C++"]
mymodule.student = {
    "name": "Ali",
    "age": 22,
    "city": "Lahore"
}

print("Function:")
mymodule.greeting("Ahmed")

print()

print("Variable:")
print(mymodule.college)

print()

print("List:")
print(mymodule.subjects)

print()

print("Dictionary:")
print(mymodule.student)

print("Student Name:", mymodule.student["name"])

print()

# Alias
mx = mymodule

print("Alias:")
print(mx.college)

print()

# Built-in Module
import platform

print("Operating System:")
print(platform.system())

print()

print("Platform Functions:")
print(dir(platform)[:10])
```

```python
# Python DateTime Practice

import datetime

# Current Date and Time
today = datetime.datetime.now()

print("Current Date and Time:")
print(today)

print()

# Individual Values
print("Year:", today.year)
print("Month:", today.month)
print("Day:", today.day)
print("Hour:", today.hour)
print("Minute:", today.minute)
print("Second:", today.second)

print()

# Weekday and Month
print("Weekday:", today.strftime("%A"))
print("Short Weekday:", today.strftime("%a"))

print("Month Name:", today.strftime("%B"))
print("Short Month:", today.strftime("%b"))

print()

# Custom Date
birthday = datetime.datetime(2005, 8, 15)

print("Birthday:")
print(birthday)

print()

# Date Formatting
print("DD/MM/YYYY :", today.strftime("%d/%m/%Y"))
print("DD-MM-YYYY :", today.strftime("%d-%m-%Y"))
print("Month Day Year :", today.strftime("%B %d, %Y"))
print("Time (24 Hour):", today.strftime("%H:%M:%S"))
print("Time (12 Hour):", today.strftime("%I:%M:%S %p"))

print()

#  ForDay Number and Week Number
print("Day of Year:", today.strftime("%j"))
print("Week Number:", today.strftime("%U"))

print()

# Complete Local Date and Time
print("Local Date and Time:")
print(today.strftime("%c"))
```
