```python
# Loop Lists

students = ["Ali", "Ahmed", "Sara", "Hamza"]

# For loop - Print each item
print("Using For Loop")
for student in students:
    print(student)

# For loop with index
print("\nUsing Range")
for i in range(len(students)):
    print(i, students[i])

# While loop
print("\nUsing While Loop")
i = 0

while i < len(students):
    print(students[i])
    i += 1

# Enumerate
print("\nUsing Enumerate")
for index, student in enumerate(students):
    print(index, student)

# Search item
print("\nSearching Student")
for student in students:
    if student == "Sara":
        print("Student Found")
```

```python
# Match Statement Practice

day = 6
grade = "B"
signal = "Green"
role = "Admin"

month = 5
workday = 2

# Check Day
match day:

    case 1:
        print("Monday")

    case 2:
        print("Tuesday")

    case 3:
        print("Wednesday")

    case 4:
        print("Thursday")

    case 5:
        print("Friday")

    case 6:
        print("Saturday")

    case 7:
        print("Sunday")

    case _:
        print("Invalid Day")


# Check Grade
match grade:

    case "A":
        print("Excellent")

    case "B":
        print("Very Good")

    case "C":
        print("Good")

    case _:
        print("Need Improvement")


# Check Traffic Signal
match signal:

    case "Red":
        print("Stop")

    case "Yellow":
        print("Ready")

    case "Green":
        print("Go")

    case _:
        print("Invalid Signal")


# Check User Role
match role:

    case "Admin":
        print("Full Access")

    case "Teacher":
        print("Teaching Access")

    case "Student":
        print("Learning Access")

    case _:
        print("Guest Access")


# Combine Multiple Values
match day:

    case 1 | 2 | 3 | 4 | 5:
        print("Weekday")

    case 6 | 7:
        print("Weekend")


# Guard Example
match workday:

    case 1 | 2 | 3 | 4 | 5 if month == 5:
        print("Working Day in May")

    case _:
        print("No Match")
```

```python
# Python *args and **kwargs

# *args Example
def numbers(*args):
    print("Numbers:", args)

numbers(10,20,30)

print()

# Sum Example
def addition(*values):

    total = 0

    for value in values:
        total += value

    print("Total:", total)

addition(5,10,15,20)

print()

# Normal + *args
def student(name,*marks):

    print("Student:", name)

    print("Marks:", marks)

student("Ali",80,85,90)

print()

# **kwargs Example
def employee(**info):

    print(info)

employee(name="Ahmed",salary=50000)

print()

# Access kwargs Values
def teacher(**data):

    print("Name:", data["name"])

    print("Subject:", data["subject"])

teacher(name="Usman",subject="Python")

print()

# Loop kwargs
def details(**record):

    for key,value in record.items():

        print(key,":",value)

details(name="Hamza",age=22,city="Lahore")

print()

# Normal + **kwargs
def person(name,**info):

    print("Name:",name)

    print(info)

person("Bilal",age=25,city="Karachi")
```

```python
# Python Scope Practice

# Global Variable
country = "Pakistan"

print("Global:", country)

print()

# Local Variable
def student():

    name = "Ali"

    print("Local:", name)

student()

print()

# Global Variable Access
city = "Lahore"

def show():

    print("City:", city)

show()

print()

# Local vs Global
name = "Ahmed"

def demo():

    name = "Usman"

    print("Local Name:", name)

demo()

print("Global Name:", name)

print()

# Global Keyword
count = 5

def update():

    global count

    count = 20

update()

print("Updated Count:", count)

print()

# Create Global Variable
def create():

    global language

    language = "Python"

create()

print(language)

print()

# Nested Scope
def outer():

    message = "Welcome"

    def inner():

        print(message)

    inner()

outer()

print()

# LEGB Rule
text = "Global Text"

def parent():

    text = "Outer Text"

    def child():

        text = "Inner Text"

        print(text)

    child()

parent()

print()

# Built-in Function
numbers = [10,20,30]

print("Length:", len(numbers))
```

```python
# Python Decorators Practice

# Basic Decorator
def decorator(function):

    def wrapper():

        print("Before Function")

        function()

        print("After Function")

    return wrapper

@decorator
def hello():

    print("Hello Python")

hello()

print()

# Decorator with Arguments
def welcome(function):

    def wrapper(name):

        print("Welcome")

        function(name)

        print("Visit Again")

    return wrapper

@welcome
def student(name):

    print("Student:", name)

student("Ali")

print()

# Decorator Returning Value
def calculate(function):

    def wrapper(a,b):

        print("Calculating...")

        result = function(a,b)

        print("Calculation Complete")

        return result

    return wrapper

@calculate
def add(a,b):

    return a+b

print("Answer:", add(10,20))

print()

# Multiple Decorators
def first(function):

    def wrapper():

        print("First Decorator")

        function()

    return wrapper

def second(function):

    def wrapper():

        print("Second Decorator")

        function()

    return wrapper

@first
@second
def message():

    print("Original Function")

message()
```

```python
# Python Lambda Practice

# Simple Lambda
add = lambda a, b: a + b
print("Addition:", add(10, 20))

print()

# Square
square = lambda x: x * x
print("Square:", square(6))

print()

# Multiply Three Numbers
multiply = lambda a, b, c: a * b * c
print("Multiply:", multiply(2, 3, 4))

print()

# Lambda Inside Function
def myFunction(n):
    return lambda a: a * n

double = myFunction(2)
triple = myFunction(3)

print("Double:", double(10))
print("Triple:", triple(10))

print()

# Lambda with map()
numbers = [1,2,3,4]

result = list(map(lambda x: x * 2, numbers))

print("Map:", result)

print()

# Lambda with filter()
numbers = [1,2,3,4,5,6]

even = list(filter(lambda x: x % 2 == 0, numbers))

print("Filter:", even)

print()

# Lambda with sorted()
names = ["Ali","Muhammad","Ahmed","Usman"]

sorted_names = sorted(names, key=lambda x: len(x))

print("Sorted:", sorted_names)
```

```python
# Python Recursion Practice

# Countdown
def countdown(n):

    if n == 0:
        print("Done")
        return

    print(n)

    countdown(n-1)

print("Countdown")
countdown(5)

print()

# Print Numbers
def numbers(n):

    if n == 6:
        return

    print(n)

    numbers(n+1)

print("Print Numbers")
numbers(1)

print()

# Factorial
def factorial(n):

    if n == 1:
        return 1

    return n * factorial(n-1)

print("Factorial:", factorial(5))

print()

# Sum
def total(n):

    if n == 0:
        return 0

    return n + total(n-1)

print("Sum:", total(5))

print()

# Power
def power(base,exp):

    if exp == 0:
        return 1

    return base * power(base,exp-1)

print("Power:", power(2,5))
```

```python
# Python Generators retrieves the next value from generator and resumes execution until next yield
def my_generator():

    yield 1
    yield 2
    yield 3

print("Simple Generator")

for value in my_generator():

    print(value)

print()

# Generator Loop
def count_up_to(n):

    count = 1

    while count <= n:

        yield count

        count += 1

print("Count Generator")

for number in count_up_to(5):

    print(number)

print()

# Using next()
def names():

    yield "Ali"
    yield "Ahmed"
    yield "Usman"

gen = names()

print("Using next()")

print(next(gen))
print(next(gen))
print(next(gen))

print()

# Generator Expression
generator = (x*x for x in range(5))

print("Generator Expression")

print(list(generator))

print()

# Sum Generator
total = sum(x*x for x in range(10))

print("Sum =",total)

print()

# Fibonacci Generator
def fibonacci():

    a,b = 0,1

    while True:

        yield a

        a,b = b,a+b

fib = fibonacci()

print("Fibonacci")

for i in range(10):

    print(next(fib))

print()

# Close Generator
def my_gen():

    try:

        yield 1
        yield 2

    finally:

        print("Generator Closed")

g = my_gen()

print(next(g))

g.close()
```
