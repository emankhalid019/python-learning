```python
# Python Class Methods - Complete Example
class Student:

    # Constructor (Runs automatically when object is created)
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Method 1 : Display Student Information
    def display_info(self):
        print(f"Name : {self.name}")
        print(f"Age  : {self.age}")

    # Method 2 : Increase Age
    def birthday(self):
        self.age += 1
        print(f"\nHappy Birthday {self.name}!")
        print(f"New Age : {self.age}")

    # Method 3 : Accept Parameters
    def add_marks(self, marks1, marks2):
        total = marks1 + marks2
        print(f"\nTotal Marks : {total}")

    # Method 4 : Uses another method
    def welcome(self):
        print(f"\nWelcome {self.name}")
        self.display_info()

    # Special Method (__str__)
    def __str__(self):
        return f"{self.name} ({self.age} years old)"

# Create Object
s1 = Student("Ali",20)

# Access Method
print("Student Information")
s1.display_info()

# Modify Property Using Method
s1.birthday()

# Method With Parameters
s1.add_marks(85,90)

# Method Calling Another Method
s1.welcome()

# __str__() Method
print("\nPrinting Object")
print(s1)

# Delete Method
# del Student.display_info

# Uncomment below line after deleting method
# s1.display_info()
```

```python
# Python Inheritance

# Parent Class
class Person:

    # Constructor
    def __init__(self, first_name, last_name):

        # Parent properties
        self.first_name = first_name
        self.last_name = last_name

    # Parent Method
    def print_name(self):
        print(f"Name : {self.first_name} {self.last_name}")

    # Parent Method
    def greet(self):
        print("Hello! Welcome.")

# Child Class
class Student(Person):

    # Child Constructor
    def __init__(self, first_name, last_name, graduation_year):

        # Call Parent Constructor
        super().__init__(first_name, last_name)

        # Child Property
        self.graduation_year = graduation_year

    # Child Method
    def welcome(self):
        print(f"Welcome {self.first_name} {self.last_name}")
        print(f"Graduation Year : {self.graduation_year}")

    # Method Overriding
    def greet(self):
        print(f"Hello Student {self.first_name}!")

# Object Creation
student1 = Student("Eman", "Khalid", 2027)

# Parent Method
student1.print_name()

# Overridden Method
student1.greet()

# Child Method
student1.welcome()

# Child Property
print("Graduation Year :", student1.graduation_year)
```

```python
# Python Polymorphism(Same method name, different behavior)

print(" Function Polymorphism ")

name = "Eman"
fruits = ["Apple", "Banana", "Mango"]
student = {
    "Name": "Eman",
    "Age": 20
}

# len() works on different data types
print("Length of Name :", len(name))
print("Number of Fruits :", len(fruits))
print("Dictionary Items :", len(student))


print("\n Inheritance + Polymorphism ")

# Parent Class
class Vehicle:

    # Constructor
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    # Parent Method
    def move(self):
        print("Vehicle is Moving")


# Child Class
class Car(Vehicle):
    pass    # Uses Parent move()


# Child Class
class Boat(Vehicle):

    # Method Overriding
    def move(self):
        print("Boat is Sailing")


# Child Class
class Plane(Vehicle):

    # Method Overriding
    def move(self):
        print("Plane is Flying")


# Object Creation
car = Car("Toyota", "Corolla")
boat = Boat("Yamaha", "Touring")
plane = Plane("Boeing", "747")


print("\n Polymorphism Output")

# Same method on different objects
for vehicle in (car, boat, plane):

    print("Brand :", vehicle.brand)
    print("Model :", vehicle.model)

    # Same method name
    vehicle.move()
```

```python
# Python Encapsulation(to protect and hide data)
class Student:

    # Constructor
    def __init__(self, name, age, grade):

        # Public Property
        self.name = name

        # Protected Property (Convention only)
        self._school = "COMSATS"

        # Private Properties
        self.__age = age
        self.__grade = grade

    # Getter Method
    # Returns private age
    def get_age(self):
        return self.__age

    # Setter Method
    # Updates age after validation
    def set_age(self, age):

        if age > 0:
            self.__age = age
        else:
            print("Age must be positive!")

    # Getter Method
    def get_grade(self):
        return self.__grade

    # Setter Method
    def set_grade(self, grade):

        if 0 <= grade <= 100:
            self.__grade = grade
        else:
            print("Grade must be between 0 and 100!")

    # Private Method
    # Only used inside class
    def __calculate_status(self):

        if self.__grade >= 50:
            return "Pass"
        else:
            return "Fail"

    # Public Method
    # Calls private method
    def show_info(self):

        print("\n Student Info ")
        print("Name :", self.name)
        print("School :", self._school)
        print("Age :", self.__age)
        print("Grade :", self.__grade)
        print("Status :", self.__calculate_status())


# Object Creation
student = Student("Eman", 19, 85)

# Public Property
print("Student Name :", student.name)

# Protected Property (Possible but not recommended)
print("School :", student._school)

# Direct access to private property will cause an error
# print(student.__age)

# Getter
print("Current Age :", student.get_age())

# Setter with valid value
student.set_age(20)

# Setter with invalid value
student.set_age(-10)

# Grade Setter
student.set_grade(92)

# Invalid Grade
student.set_grade(150)

# Display complete information
student.show_info()

# Name Mangling (Possible but NOT Recommended)
print("\nUsing Name Mangling :", student._Student__age)
```

```python
# Python Inner Classes

# Outer Class
class Car:

    # Constructor of Outer Class
    def __init__(self, brand, model):

        self.brand = brand
        self.model = model

        # Creating object of Inner Class
        self.engine = self.Engine()

        # Passing outer object to Inner Class
        self.info = self.Info(self)

    # Inner Class 1
    class Engine:

        def __init__(self):

            self.status = "OFF"

        # Start Engine
        def start(self):

            self.status = "ON"
            print("Engine Started")

        # Stop Engine
        def stop(self):

            self.status = "OFF"
            print("Engine Stopped")

    # Inner Class 2
    class Info:

        # Receive Outer Object
        def __init__(self, outer):

            self.outer = outer

        # Access Outer Class Properties
        def display(self):

            print("\nCar Information")
            print(" ")
            print("Brand :", self.outer.brand)
            print("Model :", self.outer.model)

    # Outer Class Method
    def drive(self):

        if self.engine.status == "ON":

            print(f"Driving {self.brand} {self.model}")

        else:

            print("Please Start the Engine First!")

# Object Creation
car = Car("Toyota", "Corolla")

# Display Outer Class Information
car.info.display()

print()

# Engine is OFF
car.drive()

print()

# Start Engine
car.engine.start()

print()

# Drive Car
car.drive()

print()

# Stop Engine
car.engine.stop()

print()

# Try Again
car.drive()
```

```python
#Inner Classes Examples
class Hospital:

    def __init__(self):
        self.doctor = self.Doctor()
        self.patient = self.Patient()

    class Doctor:

        def checkup(self):
            print("Doctor is checking patient.")

    class Patient:

        def treatment(self):
            print("Patient is receiving treatment.")


h = Hospital()

h.doctor.checkup()
h.patient.treatment()
```
