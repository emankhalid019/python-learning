```python
# SCIPY- Intro, Getting Started, Constants,Optimizers, Sparse Data, Graphs, Spatial Data, Matlab Arrays, Interpolation, Significance Tests


import numpy as np
import scipy
from scipy import constants
from scipy.optimize import minimize
from scipy.sparse import csr_matrix
from scipy.spatial import distance
from scipy import io
from scipy.interpolate import interp1d
from scipy import stats


## SCIPY INTRO
# SciPy = "Scientific Python"
# It is a library built on top of NumPy.
# It gives us ready-made tools for math, science, and statistics so we don't have to write everything from scratch.

print("SciPy version:", scipy.__version__)





## SCIPY GETTING STARTED

# SciPy has different parts (called subpackages) for different jobs:
#   scipy.constants   -> ready-made constants (like pi)
#   scipy.optimize    -> finding minimum/maximum values
#   scipy.sparse      -> handling data full of zeros
#   scipy.spatial     -> distance between points
#   scipy.interpolate -> guessing values between known points
#   scipy.stats       -> statistics and tests

print("We import only the part we need, e.g.: from scipy import constants")






## SCIPY CONSTANTS
# SciPy already knows many common constants, so we don't type them manually.

print("Value of Pi:", constants.pi)
print("Speed of light:", constants.speed_of_light, "meters/second")
print("1 hour in seconds:", constants.hour)
print("1 minute in seconds:", constants.minute)




## SCIPY OPTIMIZERS
# An optimizer finds the SMALLEST possible value of a function.
# Example function: f(x) = x^2 + 5

def my_function(x):
    return x**2 + 5

result = minimize(my_function, x0=0)
print("Smallest value happens at x =", result.x)
print("The smallest value of the function is:", result.fun)




## SCIPY SPARSE DATA
# Sparse data means: MOST numbers are 0, only a few are not.
# Instead of storing every 0, SciPy stores only the useful numbers.

simple_array = np.array([
    [0, 0, 3],
    [4, 0, 0],
    [0, 0, 0]
])

sparse_version = csr_matrix(simple_array)
print("Sparse version (only non-zero values stored):")
print(sparse_version)



## SCIPY GRAPHS
# A graph is just points (called "nodes") connected by lines(called "edges"). SciPy can check how nodes are connected.

from scipy.sparse.csgraph import connected_components

graph = csr_matrix([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 0]
])

n_groups, labels = connected_components(graph)
print("Number of separate connected groups:", n_groups)
print("Which group each node belongs to:", labels)


## SCIPY SPATIAL DATA
# "Spatial" means related to points/positions.
# SciPy can easily calculate the distance between two points.

point1 = (0, 0)
point2 = (3, 4)

straight_line_distance = distance.euclidean(point1, point2)
print("Distance between", point1, "and", point2, "=", straight_line_distance)



## SCIPY MATLAB ARRAYS
# MATLAB is another program used for math. SciPy lets us save and load data in MATLAB's file format (.mat).

my_array = np.array([10, 20, 30])

# Save array to a .mat file
io.savemat('my_data.mat', {"my_array": my_array})
print("Array saved to my_data.mat")

# Load it back
loaded_data = io.loadmat('my_data.mat')
print("Loaded array:", loaded_data['my_array'])


## SCIPY INTERPOLATION
# Interpolation = guessing a value that is BETWEEN two known values.
# Example: if we know values at x=1,2,3,4,5 we can guess x=2.5

x_known = [1, 2, 3, 4, 5]
y_known = [10, 20, 30, 40, 50]

guess_function = interp1d(x_known, y_known)

print("Guessed value at x=2.5:", guess_function(2.5))
print("Guessed value at x=3.5:", guess_function(3.5))



## SCIPY SIGNIFICANCE TESTS
# A significance test tells us if a difference between two groups is REAL, or just happened by random chance.

group_A = [5, 6, 7, 6, 5]
group_B = [10, 11, 12, 10, 11]

t_value, p_value = stats.ttest_ind(group_A, group_B)
print("p-value:", p_value)

if p_value < 0.05:
    print("The difference between the groups IS significant.")
else:
    print("The difference between the groups is NOT significant.")


# Topic          What it does                          Simple Example

# constants       Gives ready-made values                 constants.pi
# optimize        Finds smallest value of a function       minimize(f, x0=0)
# sparse          Stores mostly-zero data efficiently      csr_matrix(array)
# graphs          Checks how points are connected           connected_components(graph)
# spatial         Finds distance between points             distance.euclidean(p1, p2)
# io (matlab)     Saves/loads MATLAB .mat files             io.savemat() / io.loadmat()
# interpolate     Guesses values between known points        interp1d(x, y)
# stats           Tests if a difference is real              ttest_ind(groupA, groupB)
```
