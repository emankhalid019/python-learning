```python
#Global Variable(the variable that build usually outside the function)
company = "NEXABYTS"

def show_company():
    print("Company:", company)

show_company()

print("")

# Local Variable (created inside a function and only used within that function)
course = "Python"

def my_course():
    course = "Machine Learning"
    print("Inside:", course)

my_course()

print("Outside:", course)

print("")

# Create Global Variable
def student_info():
    global student #(we use keyword global if we build global var inside the function)
    student = "Eman"

student_info()

print(student)

print("")

# Change Global Variable
project = "Website"

def update_project():
    global project # we use keywoed global if we change the value of global var
    project = "AI Chatbot"

update_project()

print(project)
```

```python
#Ternary Operator(assign one value if condition is True, otherwise assign another value)

# Check voting eligibility
age = 19
print("Voting:", "Eligible" if age >= 18 else "Not Eligible")

# Check pass or fail
marks = 45
print("Result:", "Pass" if marks >= 50 else "Fail")

# Check even or odd
number = 8
print("Number:", "Even" if number % 2 == 0 else "Odd")

# Check stock
stock = 0
print("Stock:", "Available" if stock > 0 else "Out of Stock")

# Multiple conditions
day = 5
day_name = "Fri" if day == 5 else "Sat" if day == 6 else "Sun" if day == 7 else "Weekday"
print("Day:", day_name)
```

```python
# Comparison Operators

x = 15
y = 10

# Check equal values
print("Equal (==):", x == y)

# Check different values
print("Not Equal (!=):", x != y)

# Check greater value
print("Greater Than (>):", x > y)

# Check smaller value
print("Less Than (<):", x < y)

# Check greater or equal
print("Greater or Equal (>=):", x >= y)

# Check smaller or equal
print("Less or Equal (<=):", x <= y)

# Comparison with if
age = 20

if age >= 18:
    print("Eligible to Vote")
else:
    print("Not Eligible")

# Chaining comparison
number = 7

print("Chaining:", 1 < number < 10)

# Same using and
print("Using and:", 1 < number and number < 10)
```

```python
# Logical Operators
x = 7

# Both conditions must be True
print("AND:", x > 5 and x < 10)

# At least one condition must be True
print("OR:", x < 5 or x > 6)

# Reverse the result
print("NOT:", not(x > 5))

# Check voting eligibility
age = 20
citizen = True

print("Can Vote:", age >= 18 and citizen)

# Weekend check
day = "Friday"

print("Weekend:", day == "Saturday" or day == "Sunday")

# Reverse comparison
marks = 35

print("Pass Check:", not(marks >= 40))

# Using if statement
salary = 60000
experience = 3

if salary >= 50000 and experience >= 2:
    print("Eligible for Promotion")
else:
    print("Not Eligible")
```

```python
# Bitwise Operators
x = 6
y = 3

# Bitwise AND both bits must be 1
print("AND (&):", x & y)

# Bitwise OR at least one bit is 1
print("OR (|):", x | y)

# Bitwise XOR different bits become 1
print("XOR (^):", x ^ y)

# Bitwise NOT invert all bits
print("NOT (~):", ~x)

# Left Shift multiply by power of 2
print("Left Shift (<<):", x << 2)

# Right Shift divide by power of 2
print("Right Shift (>>):", x >> 1)
```

```python
# Operator Precedence

# Parentheses execute first
print("1:", (10 + 5) * 2)

# Exponentiation before multiply
print("2:", 2 ** 3 * 2)

# Multiplication before add
print("3:", 20 + 5 * 4)

# Division before subtr
print("4:", 30 - 12 / 3)

# Floor divi
print("5:", 25 // 4 + 2)

# Modulus before add
print("6:", 20 + 11 % 4)

# Comparison after arith
print("7:", 10 + 5 > 12)

# Logical AND before OR
print("8:", True or False and False)

# Parentheses change precedence
print("9:", (True or False) and False)

# Left-to-right evaluation
print("10:", 15 + 5 - 8 + 2)
```
